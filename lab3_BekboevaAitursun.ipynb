#### 2.	Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

$$
y = 2x^2 + 4x - 5
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
a, b, c = 2, 4, -5
y = a * x**2 + b * x + c

plt.plot(x, y, label='y = 2x^2 + 4x - 5', color='#FFFF00', linewidth=2)
plt.title('График квадратичной функции: y = 2x^2 + 4x - 5', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)
plt.grid(True, linestyle='--', color='#800020')
plt.legend(fontsize=10)
plt.show()
```

#### 3.	Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

$$
y = 3 \cdot \cos(3x - \frac{\pi}{6})
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
A, ω, φ = 3, 3, -np.pi / 6
y = A * np.cos(ω * x + φ)

plt.plot(x, y, label='y = 3 * cos(3x - π/6)', color='#FF1493', linewidth=2)
plt.title('График гармонической функции: y = 3 * cos(3x - π/6)', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)
plt.grid(True, linestyle='--', color='#800020')
plt.legend(fontsize=10)
plt.show()
```

#### 4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-10, 10, 100)
a, b, c = 2, 4, -5
y1 = a * x1**2 + b * x1 + c

x2 = np.linspace(0, 2 * np.pi, 100)
A, ω, φ = 3, 3, -np.pi / 6
y2 = A * np.cos(ω * x2 + φ)

fig, axs = plt.subplots(2, 1, figsize=(8, 6))

axs[0].plot(x1, y1, label='y = 2x^2 + 4x - 5', color='#FFFF00', linewidth=2)
axs[0].set_title('Квадратичная функция: y = 2x^2 + 4x - 5', fontsize=14)
axs[0].set_xlabel('Ось X', fontsize=12)
axs[0].set_ylabel('Ось Y', fontsize=12)
axs[0].grid(True, linestyle='--', color='#800020')
axs[0].legend(fontsize=10)

axs[1].plot(x2, y2, label='y = 3 * cos(3x - π/6)', color='#FF1493', linewidth=2)
axs[1].set_title('Гармоническая функция: y = 3 * cos(3x - π/6)', fontsize=14)
axs[1].set_xlabel('Ось X', fontsize=12)
axs[1].set_ylabel('Ось Y', fontsize=12)
axs[1].grid(True, linestyle='--', color='#800020')
axs[1].legend(fontsize=10)

plt.tight_layout()
plt.show()
```

#### 5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-10, 10, 100)
a, b, c = 2, 4, -5
y1 = a * x1**2 + b * x1 + c

x2 = np.linspace(0, 2 * np.pi, 100)
A, ω, φ = 3, 3, -np.pi / 6
y2 = A * np.cos(ω * x2 + φ)

plt.plot(x1, y1, label='Квадратичная функция: y = 2x^2 + 4x - 5', color='#FFFF00', linewidth=2)
plt.plot(x2, y2, label='Гармоническая функция: y = 3 * cos(3x - π/6)', color='#FF1493', linewidth=2)
plt.title('Квадратичная и Гармоническая функции', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', color='#800020')
plt.show()
```

#### 6. Построить график функции, заданной параметрически,


$$
x(t) = \cos(3t)
$$

$$
y(t) = \sin(2t)
$$

$$
t \in [0, 4\pi]
$$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi, 400)
x_t = np.cos(3 * t)
y_t = np.sin(2 * t)

plt.plot(x_t, y_t, color='#00FF7F', linewidth=2)
plt.title('Параметрический график: x(t) = cos(3t), y(t) = sin(2t)', fontsize=14)
plt.xlabel('x(t)', fontsize=12)
plt.ylabel('y(t)', fontsize=12)
plt.grid(True, linestyle='--', color='#800020')
plt.show()
```

#### 7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму

![Таблица](task7.png)

```python
import matplotlib.pyplot as plt

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7173.32]

plt.bar(years, values, color='#FFA500')
plt.title('Прожиточный минимум в Кыргызской Республике', fontsize=14)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Прожиточный минимум (сом)', fontsize=12)
plt.grid(True, linestyle='--', color='#800020')

plt.show()
```

#### 8.	Построить круговой график по данным п. 7

```python
# Построение круговой диаграммы
plt.pie(values, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Доля прожиточного минимума по годам')
plt.axis('equal')  # Для кругового отображения

# Отображение
plt.show()
```

#### 9.	Построить график произвольной каркасной поверхности.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z, color='#00BFFF', linewidth=1)

ax.set_title('Произвольная каркасная поверхность', fontsize=16, fontweight='bold')
ax.set_xlabel('X', fontsize=14)
ax.set_ylabel('Y', fontsize=14)
ax.set_zlabel('Z', fontsize=14)
ax.view_init(elev=30, azim=30)
ax.grid(True, linestyle='--', color='grey', alpha=0.7)

plt.show()
```
